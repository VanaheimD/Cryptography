Task 2: HMAC for Message Authentication
In this lab task, you will use a cryptographic key that was previously generated and stored in a file to create
and verify an HMAC for a message. This demonstrates how managed keys are used in practice for message
authentication.
1. Load the secret key from the file created in Task 1.
2. Write a function to compute an HMAC for a message using the stored key and SHA-256.
3. Write a function to verify an HMAC by recomputing it and comparing it with a provided value.
4. Input a message, generate its HMAC, and display the result.
5. Allow the user to enter an HMAC value for verification.

```python
import hmac
import hashlib

def load_key(filename):
  with open(filename, 'rb') as file:
    key = file.read()
  return key

def create_hmac(key, message):
  hmac_object = hmac.new(key, message, hashlib.sha256)
  mac = hmac_object.hexdigest()
  return mac

def verify_hmac(key, message, mac):
  compute_hmac = create_hmac(key, message)
  if compute_hmac == mac:
    return True
  else:
    return False

key = load_key('key.bin')
print("Loaded Key:", key)
message = input("Enter a message: ").encode()

hash_message = hashlib.sha256(message).hexdigest()
print("Hashed Message:", hash_message)

mac = create_hmac(key, message)
print("HMAC:", mac)

User_mac = input("Enter the HMAC value for verification: ")
if verify_hmac(key, message, User_mac):
  print("HMAC verification successful.")
else:
  print("HMAC verification failed.")
```

Task 3: Symmetric Encryption and Decryption
Your task is to implement symmetric encryption and decryption using the AES algorithm.
1. Load the secret key from the file created in Task 1.
2. Use the key to encrypt a plaintext message using AES.
3. Decrypt the ciphertext back to the original message using the same key.
4. Display the ciphertext and the decrypted message.

```python
from IPython.core.formatters import PlainTextFormatter
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
def load_key(filename):
 with open(filename, 'rb') as file:
  key = file.read()
 return key

def encrypt_message(key, plaintext):
 iv = get_random_bytes(16)
 cipher = AES.new(key, AES.MODE_CBC, iv)
 ciphertext = cipher.encrypt(pad(plaintext, AES.block_size))
 return iv, ciphertext

def decrypt_message(key, iv, ciphertext):
 decipher = AES.new(key, AES.MODE_CBC, iv)
 plaintext = unpad(decipher.decrypt(ciphertext), AES.block_size)
 return plaintext

key = load_key("key.bin")
message = input("Enter message: ").encode()
iv, ciphertext = encrypt_message(key, message)
print("Ciphertext:", ciphertext)
decrypted = decrypt_message(key, iv, ciphertext)
print("Decrypted message:", decrypted.decode())
```
