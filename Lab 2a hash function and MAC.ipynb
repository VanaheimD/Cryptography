```python
import hashlib

message = input("Enter a message:")

messgae_byte = message.encode()

hash_object = hashlib.sha256(messgae_byte)

hash_value = hash_object.hexdigest()

print("Generated SHA-256: ", hash_value)

expected_hash = input("Enter expected hash: ")

if hash_value == expected_hash:
  print("Hash matches")
else:
  print("Hash does not match")



```

```python
message1 = input("Enter a message1:")
message2 = input("Enter a message2")

hash1 = hashlib.md5(message1.encode()).hexdigest()
hash2 = hashlib.md5(message2.encode()).hexdigest()

if hash1 == hash2:
  print("Hashes are indentical")
else:
  print("Hashes are different")
```

```python
import hashlib

def generate_mac(key, message):
  data = key + message
  mac = hashlib.sha256(data).hexdigest()
  return mac

def verify_mac(key, message, mac):
  computed_mac = generate_mac(key, message)
  if computed_mac == mac:
    return True
  else:
    return False

key = input("Enter a key: ").encode()
message = input("Enter a message: ").encode()

mac = generate_mac(key, message)
print("Generated MAC: ", mac)

user_mac = input("Enter the MAC to verify: ").encode()

if verify_mac(key, message, user_mac):
  print("MAC verified")
else:
  print("MAC verification failed")
```

```python
import hmac
import hashlib

def generate_hmac(key, message):
  hmac_value = hmac.new(key, message, hashlib.sha256).hexdigest()
  return hmac_value

def verify_hmac(key, message, hmac_value):
  computed_mac = generate_hmac(key, message)
  if computed_mac == hmac_value:
    return True
  else:
    return False

key = input("Enter a key: ").encode()
message = input("Enter a message: ").encode()

hmac_value = generate_hmac(key, message)
print("Generated HMAC: ", hmac_value)

user_hmac = input("Enter the HMAC to verify: ").encode()

if verify_hmac(key, message, user_hmac):
  print("HMAC verified")
else:
  print("HMAC verification failed")
```
